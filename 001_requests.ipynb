```python
# Load env variables
from dotenv import load_dotenv

load_dotenv()
```

```python
# Create an API client
import os
from anthropic import Anthropic

client = Anthropic()
model = os.getenv('CLAUDE_MODEL')
```

```python
# Make a request
message = client.messages.create(
    model=model,
    max_tokens=100,
    messages=[
        {
            "role": "user",
            "content": "What is quantum computing? Answer in one sentence"
        }
    ]
)
```
